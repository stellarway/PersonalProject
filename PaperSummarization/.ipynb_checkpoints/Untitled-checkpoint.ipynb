```python
sourcePath = r'C:/Users/student/Documents/Papers/pdf/'
title = '0_Voice of the Customers- Mining Online Customer Reviews for Product Feature-based Rankin'
pdfPath = '{0}{1}.pdf'.format(sourcePath, title)
parsedPDF = parser.from_file(pdfPath)['content'].strip().replace('\n',' ').replace('- ','')
txtPath = pdfPath.replace('pdf','txt')
with open(txtPath, 'w', encoding = 'utf-8') as f:
    print(txtPath)
    f.write(parsedPDF)
    f.close()
```

```python
a = re.sub(r'\s*8\s*References.*','',parsedPDF)
```

```python
token1 = word_tokenize(a)
print(token1)
```

```python
taggedToken = pos_tag(token1)
from pprint import pprint
pprint(taggedToken)
```

```python
neToken=ne_chunk(taggedToken)
print(neToken)
```
